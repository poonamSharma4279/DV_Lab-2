```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate

# Load dataset from github
df = "https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/2_TwoNum.csv"
df_2 = pd.read_csv(df, usecols=['GrLivArea', 'SalePrice'])

df_2
```

```python

# For 1st plot
fig, ax1 = plt.subplots()
ax1.hist(df_2['GrLivArea'], bins=50, color='#69b3a2', edgecolor='#e9ecef', alpha=0.9)
ax1.set_xlabel('area')
ax1.set_ylabel('Frequency')
ax1.set_title('Ground living area distribution')
```

```python

# For 2nd plot
fig, ax2 = plt.subplots()
ax2.hist(df_2['SalePrice']/1000, bins=50, color='#69b3a2', edgecolor='#e9ecef', alpha=0.9)
ax2.set_xlabel('Sale price (k$)')
ax2.set_ylabel('Frequency')
ax2.set_title('Sale price distribution')

# Combine the plots
fig.tight_layout()
plt.show()
```

```python


fig, ax = plt.subplots()
ax.scatter(df_2['GrLivArea'], df_2['SalePrice']/1000, color='#69b3a2', alpha=0.8)
ax.set_xlabel('Ground living area')
ax.set_ylabel('Sale price (k$)')
ax.set_title('Ground living area partially explains sale price of apartments')

# Show the plot
fig.tight_layout()
plt.show()
```

```python

sns.scatterplot(x='GrLivArea', y='SalePrice', data=df_2, color='#69b3a2', alpha=0.8)
plt.title("Linear regression", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
sns.regplot(x='GrLivArea', y='SalePrice', data=df_2, scatter=False, color="black", line_kws={'alpha':0.8,'lw':0.5,'ls':'--'})
plt.show()
```

```python

sns.scatterplot(x='GrLivArea', y='SalePrice', data=df_2, color='#69b3a2', alpha=0.8)
plt.title("Loess", size=12)
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
sns.regplot(x='GrLivArea', y='SalePrice', data=df_2, scatter=False, lowess=True, color="black", line_kws={'alpha':0.8,'lw':0.5,'ls':'--'})

plt.show()
```

```python
plt.figure(figsize=(20,5))

plt.figure(figsize=(6,6))
sns.kdeplot(x=df_2["GrLivArea"], y=df_2["SalePrice"],color="#69b3a2")
plt.show()


plt.figure(figsize=(6,6))
sns.kdeplot(x=df_2["GrLivArea"], y=df_2["SalePrice"], cmap="Greens", shade=True, thresh=0)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.hexbin(df_2["SalePrice"], df_2["GrLivArea"], cmap ='Greens')
plt.show()

plt.figure(figsize=(6,6))
plt.hexbin(df_2['GrLivArea'], df_2['SalePrice']/1000, gridsize=30, cmap='viridis', mincnt=1)
plt.show()

plt.figure(figsize=(6,6))
plt.hist2d(df_2['GrLivArea'], df_2['SalePrice']/1000, bins=30, cmap='viridis', cmin=1)
plt.show()
```
