```python
import pandas as pd
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/11_SevCatOneNumNestedOneObsPerGroup.csv", sep=';')

# Missing values
df.dropna(inplace=True)
df.head()
df.columns=["Continent", "Region", "Country", "Pop"]
```

```python
df = df[df['Pop'].notna()]
df = df.sort_values(by='Pop')
df = df.tail(50)
df = df.sort_values(by=['Continent', 'Pop'])
df['Country'] = df['Country'].astype('category')
df['Pop'] = df['Pop'] / 1000000

plt.figure(figsize=(12,6))
sns.scatterplot(x='Country', y='Pop', hue='Continent', data=df)
sns.despine(left=True, bottom=True)
plt.xlabel('')
plt.ylabel('Population (M)')
```
