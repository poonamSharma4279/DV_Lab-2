```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px

df = pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/3_TwoNumOrdered.csv", sep=" ")
df["date"] = pd.to_datetime(df["date"])
df
```

```python
plt.figure(figsize=(6,6))
plt.plot(df['date'], df['value'],color='red')
plt.xlabel('date')
plt.ylabel('bitcoin price ($)')


plt.figure(figsize=(6,6))
plt.fill_between(df['date'], df['value'], color='red')
plt.xlabel('date')
plt.ylabel('bitcoin price ($)')
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['date'].tail(10), df['value'].tail(10), linestyle='-', marker='o', color='red')
plt.show()
```

```python
y1 = df.tail(300)[df.tail(300)['value']<7500]
y2 = df.tail(300)[df.tail(300)['value']>=7500]

plt.figure(figsize=(6,6))
plt.plot(y1['date'], y1['value'], color='red')
plt.plot(y2['date'], y2['value'], color='black')
plt.show()
```
