```python
import pandas as pd
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/11_SevCatOneNumNestedOneObsPerGroup.csv", sep=';')
df
```

Preprocessing of Data

```python
df= df.dropna(subset=['value'])
```

```python
df
```

```python
df = df.sort_values(by='value').tail(20)
f = plt.figure(figsize=(8, 8))

# Plot the horizontal lines
plt.hlines(y=range(1, len(df.index) + 1), xmin=0, xmax=df['value'], color='black')

# Plot the df points
plt.scatter(df['value'], range(1, len(df.index) + 1), color='red', s=175)
plt.yticks(range(1, len(df.index) + 1), df['key'])
plt.xlabel('value')
plt.show()
```
