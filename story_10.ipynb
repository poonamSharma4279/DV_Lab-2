```python
import pandas as pd
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/9_OneNumSevCatSubgroupOneObs.csv",  sep=",")
df.head()
```

```python
with4 = df.groupby('Country')['Country'].count().reset_index(name='n')
with4 = with4[with4['n'] == 4]
df = df[df['Country'].isin(with4['Country'])]
df['Country'] = df['Country'].astype('category')
df['TIME'] = df['TIME'].astype('category')
df['TIME'].cat.reorder_categories(df['TIME'].cat.categories, inplace=True)
plt.figure(figsize=(20,8))
sns.barplot(x='Value', y='Country', hue='TIME', data=df, dodge=True)
sns.despine(left=True, bottom=True)
plt.xlabel('')
plt.ylabel('Gender wage gap (%)')
plt.legend(title='')
plt.show()
```

```python
with4 = df.groupby('Country')['Country'].count().reset_index(name='n')
with4 = with4[with4['n'] == 4]
df = df[df['Country'].isin(with4['Country'])]
df['Country'] = df['Country'].astype('category')
df['TIME'] = df['TIME'].astype('category')
df['TIME'].cat.reorder_categories(df['TIME'].cat.categories, inplace=True)
df['Country'] = df.groupby('Country')['Value'].transform(lambda x: x.mean())
df.sort_values(by=['Country'],inplace=True)

plt.figure(figsize=(8,6))
sns.barplot(x='Value', y='TIME', hue='Country', data=df, dodge=True)
sns.despine(left=True, bottom=True)
plt.xlabel('')
plt.ylabel('Gender wage gap (%)')
plt.legend(title='')
```
